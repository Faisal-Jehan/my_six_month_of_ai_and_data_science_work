# 1.Min Max Scaling

```python
from sklearn.preprocessing import MinMaxScaler

import pandas as pd
```

```python
data = {'Aven numbers': [10,20,30,40,50]}
df = pd.DataFrame(data)
df.head()

```

```python
# Min.Max Scaling 
scaler= MinMaxScaler()
df['Aven numbers scaled']= scaler.fit_transform(df[['Aven numbers']])
df.head()
```

# 2.Standard Scalar or Z-score normalization


```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Sample data
data = {'Aven numbers': [10,20,30,40,50]}
df = pd.DataFrame(data)
df.head()

```

```python
# Scale the data using Standard Scalar
scaler= StandardScaler()
df['Aven numbers scaled']= scaler.fit_transform(df[['Aven numbers']])
df.head()
```

# 3. Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
import pandas as pd

# Sample data
data = {'Aven numbers': [10,20,30,1000,50]}
df = pd.DataFrame(data)
df.head()

```

```python
# Scale the data using Robust Scalar
scaler= RobustScaler()
df['Aven numbers scaled']= scaler.fit_transform(df[['Aven numbers']])
df.head()
```

# 4.Logrithmic scaling/ Normalization

```python
import numpy as np
```

```python
# random data with outliers
data= {'awen_numbers': [10000,20000,30000,1000000,50000]}
df=pd.DataFrame(data)

# Log Transform

df['awen_numbers_log'] = np.log(df['awen_numbers'])
df['awen_numbers_log2'] = np.log2(df['awen_numbers'])
df['awen_numbers_log10'] = np.log10(df['awen_numbers'])

df.head()
```

```python
import seaborn as sns
sns.boxplot(df, x='awen_numbers_log10')
```

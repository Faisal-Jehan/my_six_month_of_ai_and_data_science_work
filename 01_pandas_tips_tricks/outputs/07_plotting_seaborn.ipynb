# Seaborn using plotting

```python
# import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# important things for Plotting according to graphics

2. Types of plot?
3. x axis
4. y-axis
5. grouping (hue/color) based on categorical variable

7. color palletes of the plot
8. style of the plot
9. how do you interpret

```python
# Data 
df= sns.load_dataset('Titanic')
```

```python
df.info()
```

```python
# scatter plot
sns.scatterplot(df, x='age', y='fare', hue='sex')
```

```python
# style of a plot 
sns.set_style('darkgrid')

# scatter plot size of figure size
plt.figure(figsize=(8,4)) # a4 page size 16:9

# change the scatter point colors
pallette={'male':'green','female':'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=pallette)

# labels 
plt.xlabel('age')
plt.ylabel('Fare')
plt.title('Scatter plot of age vs Fare')
```

```python
# style of a plot 
sns.set_theme(style='darkgrid', font_scale=1.2, font='calibri')

# scatter plot size of figure size
plt.figure(figsize=(16,4)) # a4 page size 16:9

# change the scatter point colors
pallette={'male':'green','female':'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', size='fare', palette=pallette)

# labels 
plt.xlabel('age')
plt.ylabel('Fare')
plt.title('Scatter plot of age vs Fare')
```

# Interpretation:
1. We have Age of passengers at x-axis and fare at y-axis
2. The scatter plot shows the relationship between age and fare of titanic passengers.
3. We further grouped the passenger by their genders such as male(Green points) and female (Red points)
4. Two male passengers gave the highest fare in titanic, and they might be travelling first class.
5. The oldest person at titanic is also a male 80 years of age as show in scatter plot. (As shown in figure 1).

# BOX PLOT

```python
df['class'].unique()
```

```python
# style of a plot 
sns.set_theme(style='darkgrid', font_scale=0.7, font='Times New Roman')

# scatter plot size of figure size
plt.figure(figsize=(8,4)) # a4 page size 16:9

# change the scatter point colors
pallette={'First':'green','Second':'red', 'Third': 'blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=pallette)

# labels 
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of age vs Fare')

# Change legend position and title
plt.legend(loc='lower left', title='Class Name')
```

```python
# style of a plot 
sns.set_theme(style='dark', font_scale=0.7, font='Times New Roman')

# scatter plot size of figure size
plt.figure(figsize=(8,4)) # a4 page size 16:9

# change the scatter point colors
pallette={'First':'green','Second':'red', 'Third': 'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=pallette, capsize=0.1)

# labels 
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxplot of Age, based on Gender and Class Name of Passengerss')

# Change legend position and title
plt.legend(loc='lower left', title='Class Name')

# Change the x tick labels
plt.xticks(ticks=[0,1], labels=["Male", "Female"],)
```

```python
# style of a plot 
sns.set_theme(style='dark', font_scale=0.7, font='Times New Roman')

# scatter plot size of figure size
plt.figure(figsize=(8,4)) # a4 page size 16:9

# change the scatter point colors
pallette={'First':'green','Second':'red', 'Third': 'blue'}
sns.violinplot(data=df, x='sex', y='age', hue='class', palette=pallette, )

# labels 
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Boxplot of Age, based on Gender and Class Name of Passengerss')

# Change legend position and title
plt.legend(loc='lower left', title='Class Name')

# Change the x tick labels
plt.xticks(ticks=[0,1], labels=["Male", "Female"],)
```

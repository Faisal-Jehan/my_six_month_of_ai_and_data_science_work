```python
# import libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
names=["faisal",'talha','ahmad']
heights= [5.8,7,5.6]
```

```python
# Create a plot (line chart)
fig=px.bar(x=names, y=heights, title="The Heights of the Students")

# Show the fig
fig.show()

# save the plot in png
fig.write_image("./output/plotly.png")

# save as standaone html
fig.write_html("./output/01_barplot.html")
```

# Plots on titanic Dataset

```python
# import dataset
df=sns.load_dataset('titanic')

# Check the head of the dataset
df.head()
```

```python
# scatter plot
df.dropna(inplace=True)
fig=px.scatter(df, x="age", y='fare', color='sex', size='fare', title="Plot of Titanic Dataset")
fig.show()

fig.write_image("./output/02_scatterplot.png", format='png', width=800, height=600, scale=-3) #how to save in 300 DPI
```

```python
import plotly.graph_objects as go

# Create a simple line chart 
fig=go.Figure(data=go.Scatter(x=[1,2,3,4,],y=[10,11,12,13]))
fig.show()
```

```python
# Customize 

fig.update_layout(
    title ="line plot of dummy data",
    xaxis=dict(title="Numbers"),
    yaxis=dict(title="values"), 
    showlegend=True,)
```

```python
# Heatmap in plotly

data= np.random.rand(10,10)

heatmap=go.Heatmap(z=data)
fig=go.Figure(data=[heatmap])
fig.show()
```

```python
df[['age','fare']].corr()
```

```python
# Heatmap on corr of titanic dataset of age and fare using plotly

fig=px.imshow(df[['age','fare']].corr())
fig.show()
```
